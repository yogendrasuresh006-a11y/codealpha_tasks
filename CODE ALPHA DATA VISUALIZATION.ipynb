```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel(
    r"C:\Users\Suresh\Downloads\sample_-_superstore.xls",
    engine="xlrd"
)
```

# Superstore Sales Data Visualization

This project uses data visualization techniques to analyze sales, profit, regional performance, and customer purchasing patterns.

The goal is to transform raw data into meaningful visual insights that support business decision-making.

## Visualization 1: Sales by Category

```python
category_sales = df.groupby("Category")["Sales"].sum()

plt.figure(figsize=(8,5))

sns.barplot(
    x=category_sales.index,
    y=category_sales.values
)

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")

plt.show()
```

### Insight

Technology generated the highest sales among all categories, indicating strong customer demand for technology products.

## Visualization 2: Profit by Category

```python
category_profit = df.groupby("Category")["Profit"].sum()

plt.figure(figsize=(8,5))

sns.barplot(
    x=category_profit.index,
    y=category_profit.values
)

plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Total Profit")

plt.show()
```

### Insight

Technology generated the highest profit among all categories, while Furniture generated the lowest profit.

Although Furniture achieved strong sales, its profitability was significantly lower than Technology and Office Supplies.

## Visualization 3: Regional Sales Distribution

```python
region_sales = df.groupby("Region")["Sales"].sum()

plt.figure(figsize=(7,7))

plt.pie(
    region_sales,
    labels=region_sales.index,
    autopct="%1.1f%%"
)

plt.title("Regional Sales Distribution")

plt.show()
```

### Insight

The West region contributed the highest share of total sales, followed by the East region.

The South region generated the lowest sales among all regions.

This indicates that the company's strongest market presence is in the West region.

## Visualization 4: Top 10 Selling Sub-Categories

```python
top10 = (
    df.groupby("Sub-Category")["Sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top10.index,
    y=top10.values
)

plt.xticks(rotation=45)

plt.title("Top 10 Selling Sub-Categories")
plt.xlabel("Sub-Category")
plt.ylabel("Total Sales")

plt.show()
```

### Insight

Phones generated the highest sales among all sub-categories, closely followed by Chairs.

Storage, Tables, and Binders also contributed significantly to total revenue.

These sub-categories represent the company's key revenue drivers and should be prioritized in inventory and marketing strategies.

## Visualization 5: Discount vs Profit

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x="Discount",
    y="Profit"
)

plt.title("Discount vs Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")

plt.show()
```

### Insight

The scatter plot shows a negative relationship between Discount and Profit.

As discounts increase, profit tends to decrease.

Several orders with very high discounts resulted in significant losses, indicating that excessive discounting can negatively impact business profitability.

This visualization supports the findings from the correlation analysis performed during EDA.

## Visualization 6: Correlation Heatmap

```python
numeric_df = df.select_dtypes(include="number")

plt.figure(figsize=(8,5))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()
```

### Insight

The heatmap shows the relationships among numerical variables in the dataset.

Discount has a negative correlation with Profit, while Sales has a positive relationship with Profit.

These relationships help identify factors that influence business performance.
