# Exploratory Data Analysis (EDA)

## Business Question 1:
Which category generates the highest sales?

```python
df.groupby("Category")["Sales"].sum()
```

### Observation

Technology generated the highest sales with total sales of 836,154.03.

Furniture generated sales of 741,999.80, while Office Supplies generated sales of 719,047.03.

Therefore, Technology is the best-performing category in terms of revenue.

## Business Question 2:
Which region generates the highest sales?

```python
df.groupby("Region")["Sales"].sum()
```

### Observation

The West region generated the highest sales with total sales of 725,457.82.

The East region generated sales of 678,781.24, followed by the Central region with 501,239.89.

The South region recorded the lowest sales at 391,721.91.

Therefore, the West region is the best-performing region in terms of revenue generation.

## Business Question 3:
Which category generates the highest profit?

```python
df.groupby("Category")["Profit"].sum()
```

### Observation

Technology generated the highest profit with a total profit of 145,454.95.

Office Supplies generated a profit of 122,490.80.

Furniture generated the lowest profit at 18,451.27.

Although Furniture achieved high sales, its profit was significantly lower compared to Technology and Office Supplies.

Therefore, Technology is the most profitable category in the dataset.

## Business Question 4:
Do discounts affect profit?

```python
df[["Discount", "Profit"]].corr()
```

### Observation

The correlation between Discount and Profit is -0.219.

This indicates a negative relationship between discounts and profit.

As discounts increase, profit tends to decrease.

Therefore, the hypothesis that higher discounts reduce profit is supported by the data.

## Business Question 5:
Which sub-categories generate the highest sales?

```python
df.groupby("Sub-Category")["Sales"].sum() \
  .sort_values(ascending=False) \
  .head(10)
```

### Observation

Phones generated the highest sales with total sales of 330,007.05.

Chairs closely followed with sales of 328,449.10.

Storage, Tables, and Binders were also among the top-performing sub-categories.

The top 5 revenue-generating sub-categories are:
1. Phones
2. Chairs
3. Storage
4. Tables
5. Binders

This indicates that these product groups contribute significantly to the company's overall revenue.

## Business Question 6:
Which orders generated the highest losses?

```python
df.nsmallest(10, "Profit")[["Product Name", "Category", "Sales", "Discount", "Profit"]]
```

### Observation

The analysis identified several orders that generated significant losses.

The largest loss was caused by the product "Cubify CubeX 3D Printer Double Head Print", which resulted in a loss of 6,599.98.

Most loss-making products belong to the Technology and Office Supplies categories.

A common pattern among these products is the presence of very high discounts (50% to 80%).

This suggests that excessive discounting has a strong negative impact on profitability and should be carefully managed.

Therefore, the company's discount strategy should be reviewed to reduce financial losses and improve overall profit margins.

## Business Question 7:
Which orders generated the highest profits?

```python
df.nlargest(10, "Profit")[["Product Name", "Category", "Sales", "Discount", "Profit"]]
```

### Observation

The product "Canon imageCLASS 2200 Advanced Copier" generated the highest profit of 8,399.98.

Most of the highest-profit products belong to the Technology category, indicating strong profitability in this segment.

Several high-profit products were sold with little or no discount, suggesting that maintaining healthy pricing contributes significantly to profitability.

Office Supplies also contributed several highly profitable products, particularly binding systems and office equipment.

These products should be prioritized in marketing, inventory planning, and sales strategies to maximize revenue and profit.
